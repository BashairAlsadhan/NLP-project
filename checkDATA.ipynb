```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Check
new_data = pd.read_csv('combined_data.csv')

print(new_data.SpeakerDialect.unique())
print(new_data.Category.unique())

# Create a pivot table to count occurrences of each category per dialect
dialect_category_matrix = new_data.pivot_table(index="SpeakerDialect", columns="Category", aggfunc="size", fill_value=0)

# Print the matrix
print(dialect_category_matrix)

# Visualize the data using a heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(dialect_category_matrix, annot=True, fmt="d", cmap="Blues", linewidths=0.5)
plt.xlabel("Category")
plt.ylabel("Dialect")
plt.title("Number of Categories per Dialect")
plt.xticks(rotation=45)
plt.show()
```
